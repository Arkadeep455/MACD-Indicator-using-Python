```python
import yfinance as yf
```

```python
aapl = yf.Ticker('AAPL')
data= aapl.history(interval='1h', period='3mo')
```

```python
data
```

```python
data['EMA12']=data['Close'].ewm(span=12, adjust=False).mean()
```

```python
data['EMA26']=data['Close'].ewm(span=26, adjust=False).mean()
```

```python
data
```

```python
data['MACD']=data['EMA12']-data['EMA26']
```

```python
data
```

```python
data['Signal_line']=data['MACD'].ewm(span=9, adjust=False).mean()
```

```python
data
```

```python
last_row=data.iloc[-1]
second_last_row=data.iloc[-2]
```

```python
print(f"last_row MACD: {last_row['MACD']}")
print(f"last_row signal line: {last_row['Signal_line']}")
print(f"second_last_row MACD: {second_last_row['MACD']}")
print(f"second_last_row signal line: {second_last_row['Signal_line']}")
```

```python
if second_last_row['MACD'] > second_last_row['Signal_line'] and last_row['MACD']<last_row['Signal_line']:
  print('cross over')

if second_last_row['MACD'] < second_last_row['Signal_line'] and last_row['MACD']>last_row['Signal_line']:
  print('cross above')

else:
  print('no cross over')
```
